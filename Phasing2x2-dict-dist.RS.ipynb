```python
# Adjust fourier error threshold
fourier_error_final_s = copy(fourier_error_final[:])
fmax = 0.190
s = [s for s,f in zip(range(N_rec),fourier_error_final[:]) if f < fmax]
fourier_error_final_s.sort()
plot(fourier_error_final_s)
axhline(fmax)
fs_max = abs(fourier_space_final[s,:,:]).max()
fs_levels = [0.175,0.179,0.19]
for fs_level in fs_levels: axhline(fs_level,c="black")
```

```python
# Output PNGs in classed folders
class_1 = fourier_error_final < fs_levels[0]
class_2 = fourier_error_final < fs_levels[1] * (class_1 == False)
class_3 = fourier_error_final < fs_levels[2] * (class_1 == False) * (class_2 == False)
class_4 = (class_1 == False) * (class_2 == False) * (class_3 == False)
#plot(class_1)
vmin = 0.
vmax = abs(real_space_final).max()
for i,c in enumerate([class_1,class_2,class_3,class_4]):
    l = "class_%i" % i
    os.system("mkdir -p %s" % l)
    os.system("rm %s/*.png" % l)
    for k in arange(len(c))[c]:
        imsave("%s/img%04i.png" % (l,k),abs(real_space_final[k,:,:]),vmin=vmin,vmax=vmax)
```
