```python
import spimage,h5py
```

```python
area_init_def = 0.020
area_final_def = 0.018
data_folder = "/home/higsur-a/data/extraction_data/r0184_20150210_1158_extraction"

D = {}

D["2995"] = {"filename":data_folder+"/2995_2.h5",
             "area_init":area_init_def,
             "area_final":area_final_def,
             "fourier_error_max":0.25}
D["2988"] = {"filename":"./test_data/good_pattern/2088_2.h5",
             "area_init":area_init_def,
             "area_final":area_final_def,
             "fourier_error_max":0.25},
D["49"]   = {"filename":data_folder+"/49_2.h5",
             "area_init":area_final_def,
             "area_final":area_final_def,
             "fourier_error_max":0.25}
```

```python
identifier = "49"
S = D[identifier]
filename = S["filename"]
filename_cxi = "./rec_%s.cxi" % identifier
img = spimage.sp_image_read(filename,0)
I = abs(img.image)
M = array(img.mask,dtype="bool")
spimage.sp_image_free(img)
```

```python
S["area_final"]
S.keys()
```

```python
I,M = spimage.downsample(I,2,mask=M)
```

```python
imshow(log10(I)*log10(M*10))
```

```python
R = spimage.Reconstructor()
```

```python
R.set_intensities(I,shifted=False)
```

```python
R.set_mask(M,shifted=False)
```

```python
R.set_number_of_iterations(4000)
```

```python
R.set_number_of_outputs(100)
```

```python
R.set_initial_support(area=S["area_init"])
```

```python
R.set_phasing_algorithm("raar",beta_init=0.9,beta_final=0.9,constraints=["enforce_positivity"],number_of_iterations=2000)
R.append_phasing_algorithm("raar",beta_init=0.9,beta_final=0.9,constraints=["enforce_positivity"],number_of_iterations=1000)
R.append_phasing_algorithm("er",constraints=["enforce_positivity"],number_of_iterations=1000)
```

```python
R.set_support_algorithm("static",update_period=2000,number_of_iterations=2000,center_image=False)
R.append_support_algorithm("area",blur_init=6.,blur_final=1.,area_init=S["area_init"],area_final=S["area_final"],update_period=100,number_of_iterations=1000,center_image=False)
R.append_support_algorithm("static",number_of_iterations=1000,center_image=False)
```

```python
O = R.reconstruct()
```

```python
#O.keys()
```

```python
rs = O["real_space"]
fs = O["fourier_space"]
su = O["support"]
```

```python
for i in linspace(0,fs.shape[0]-1,5):
    fig = figure(figsize=(9,3))
    ax0 = fig.add_subplot(1,3,1)
    ax0.imshow(abs(rs[int(i),:,:])*log10(10*su[int(i),:,:]))
    ax0 = fig.add_subplot(1,3,2)
    ax0.imshow(numpy.angle(rs[int(i),:,:])*log10(10*su[int(i),:,:]))
    ax1 = fig.add_subplot(1,3,3)
    ax1.imshow(log10(abs(fs[int(i),:,:])**2),vmin=-1,vmax=6)
```

```python
N = 10
```

```python
O_N = R.reconstruct_loop(N)
```

```python
W = spimage.CXIWriter(filename_cxi,N)
W.write_stack(O_N)
W.close()
```

```python
with h5py.File(filename_cxi,"r") as f:
    fourier_space_final = array(f["fourier_space_final"])
    real_space_final = array(f["real_space_final"])
    support_final = array(f["support_final"])
    fourier_error_final = array(f["scores_final"]["fourier_error"])
    real_error_final = array(f["scores_final"]["real_error"])
```
