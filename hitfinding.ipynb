```python
import h5py
import scipy
from scipy.optimize import leastsq
from scipy.special import erf,erfinv
import os,sys
```

```python
root_dir = "/Users/hantke/mnt/"
```

```python
rDicts = {
            "r0179":{"injection":"Paramecium bursaria chlorella virus","reference":"Xe"},
            "r0180":{"injection":"Paramecium bursaria chlorella virus","reference":"Xe"},
            "r0181":{"injection":"Paramecium bursaria chlorella virus","reference":"Xe"},
            "r0182":{"injection":"Paramecium bursaria chlorella virus","reference":"Xe"},
            "r0183":{"injection":"Paramecium bursaria chlorella virus","reference":"Xe"},
            "r0184":{"injection":"Paramecium bursaria chlorella virus","reference":"Xe"},
            "r0185":{"injection":"Paramecium bursaria chlorella virus","reference":"Xe"},
            "r0186":{"injection":"Paramecium bursaria chlorella virus","reference":"Xe"},
            "r0187":{"injection":"Paramecium bursaria chlorella virus","reference":"Xe"},
            "r0188":{"injection":"Paramecium bursaria chlorella virus","reference":"Xe"},
            "r0189":{"injection":"Paramecium bursaria chlorella virus","reference":"Xe"},
          }


def plot_run(run,rDict):
    inj = rDict["injection"]
    ref = rDict["reference"]
    data_filename = root_dir + "/test_%s/amoc6914-%s.cxi" % (run,run)
    cache_filename = root_dir + "/test_%s/hitfinding.h5" % (run)
    if True or not os.path.exists(cache_filename):
        with h5py.File(data_filename,"r") as f:
            nPeaks = array(f["/cheetah/event_data/nPeaks"])
            root_group = "/entry_1/image_1/detector_corrected"
            if "tags" in f["/entry_1/image_1/detector_corrected"].keys():
                root_group = "/entry_1/image_1/detector_corrected"
            else:
                root_group = "/entry_1/image_1"
            hit = array(f[root_group+"/tags"][0],dtype="bool")
            i = nPeaks.argmax()
            image = array(f["/entry_1/image_1/data"][i,:,:])
            blank = (hit==False)
            blank[20:] = False
            bg = array(f["/entry_1/image_1/data"][blank,:,:]).mean(0)
            fiducial = array(f["/LCLS/fiducial"])
        with h5py.File(cache_filename,"w") as f:
            f["/nPeaks"] = nPeaks
            f["/hit"] = hit
            f["/image"] = image
            f["/blank"] = blank
            f["/bg"] = bg
            f["/fiducial"] = fiducial
    else:
        with h5py.File(cache_filename,"r") as f:
            nPeaks = array(f["/nPeaks"])
            hit = array(f["/hit"])
            i = nPeaks.argmax()
            image = array(f["/image"])
            #blank = array(f["/blank"])
            bg = array(f["/bg"])
            fiducial = array(f["/fiducial"])
    P = linspace(0.,1.,20)
    s = nPeaks.argsort()[::-1]
    hit_s = hit[s]
    nPeaks_s = nPeaks[s]
    P_N = linspace(0.,1.,len(hit_s))
    hit_fraction = zeros(len(P))
    for i,p in zip(range(len(P)),P):
        M = abs(P_N-p) < 0.05
        hit_fraction[i] = hit_s[M].sum()/float(M.sum())
    fit_function = lambda x,v: (1-erf((x-v[0])/v[1]))*0.5
    fit_function_i = lambda ff,v: v[0] + v[1]*erfinv(1-ff/0.5)
    err = lambda v: abs(hit_fraction-fit_function(P,v))**2
    v0 = [0.5,0.1,1.]
    v, success = leastsq(err, v0, maxfev=1000)
    xplot = linspace(0.,1.,100)
    fit = fit_function(xplot,v)
    hitratio = v[0]
    hitratio_strong = max([fit_function_i(0.99,v),0.])
        
    fig = figure(figsize=(18,4))
    if ref.lower() != "none":
        fig.suptitle(run + " - %s + %s - %d hits / %d shots" % (inj, ref, hit.sum(), len(hit)))
    else:
        fig.suptitle(run + " - %s - %d hits / %d shots" % (inj, hit.sum(), len(hit)))
    ax0 = fig.add_subplot(141)
    ax0.set_title("Statistics of hit finding")
    ax0.plot(P,hit_fraction,".",c="red")
    ax0.plot(xplot,fit)
    if hitratio > 0:
        nPeaks_min = nPeaks_s[int(round(hitratio*(len(nPeaks)-1)))]
        nPeaks_min_strong = nPeaks_s[int(round(hitratio_strong*len(nPeaks)))-1]
        ax0.axvline(hitratio,c="black")
        ax0.axvline(hitratio_strong,c="blue")
        ax0.text(hitratio+0.025,0.75,"HITS")
        ax0.text(hitratio+0.025,0.5,"hit ratio = %i %%" % int(round(100*hitratio)))
        ax0.text(hitratio+0.025,0.6,"> %i\nlit pixels" % nPeaks_min)
        if hitratio_strong > 0.:
            if hitratio_strong > 0.1:
                pos = hitratio_strong-0.025
                ha = "right"
            else:
                pos = hitratio + 0.5
                ha = "left"
            ax0.text(pos,0.75,"STRONG HITS",ha=ha,color="blue")
            ax0.text(pos,0.6,"> %i\nlit pixels" % nPeaks_min_strong,ha=ha,color="blue")
            ax0.text(pos,0.5,"hit ratio = %i %%" % int(round(100*hitratio_strong)),ha=ha,color="blue")
    ax0.set_xlabel("data fraction")
    ax0.set_ylabel("hit fraction")
    ax0.set_ylim(-0.1,1.1)
    ax0.set_xlim(0,1.)
    ax1 = fig.add_subplot(142)
    ax1.set_title("Strongest hit")
    image[image<50.] = 50.
    ax1.imshow(log10(image),vmin=log10(50.),vmax=log10(10000),cmap=cm.gnuplot)
    ax2 = fig.add_subplot(143)
    bg[bg<50.] = 50.
    ax2.set_title("Background average")
    ax2.imshow(log10(bg),vmin=log10(50.),vmax=log10(10000),cmap=cm.gnuplot)
    ax3 = fig.add_subplot(144)
    ax3.set_title("Occurence of hits")
    ax3.plot(fiducial-fiducial[0],hit,".")
    ax3.set_xlabel("Fiducial")
    ax3.set_ylabel("Hit flag")
    ax3.set_ylim(-0.2,1.2)
    savefig("./%s_hitfinding.png" % run)
```
